```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
data = {}
for c in ['nyc', 'la', 'london']:
    file = '../../Data/data_'+c+'.csv'
    data[c] = pd.read_csv(file)
    data[c] = data[c][data[c]['disadvantaged']==True]
```

```python
plt.rcParams["figure.figsize"] = [5,4]
fig = plt.figure()

ax = sns.distplot(data['nyc']['index_change'], kde_kws={"color": "k"},rug_kws={"color": "k"}, hist_kws={"color":"gray"})

plt.ylabel('Probability Density',fontsize=14)
plt.xlabel('Gentrification Score',fontsize=14)
plt.title('New York',fontsize=15)

plt.ylim(0,0.055)
plt.yticks(np.arange(0, 0.06, 0.01),fontsize=12)

plt.xlim(-60,60)
plt.xticks(np.arange(-50, 55, 25),fontsize=12)

plt.tight_layout()
plt.show()
#fig.savefig('nyc_dist.eps', format='eps', dpi=1000)
```
