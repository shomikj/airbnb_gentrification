```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```python
gentrifying_map = {
    'williamsburg_greenpoint': [11211, 11222, 11249, 11206],
    'central_harlem': [10026, 10027, 10030, 10037, 10039],
    'lower_east_side_chinatown': [10002, 10013, 10003, 10009],
    'bushwick': [11206, 11221, 11237],
    'east_harlem': [10029, 10035],
    'morningside_heights': [10025, 10027, 10031, 10032, 10039],
    'bedford_stuyvesant': [11206, 11216, 11221],
    'crown_heights_north_prospect_heights': [11238, 11216, 11213, 11233, 11205, 11238],
    'washington_heights_inwood': [10032, 10033, 10040, 11096, 11559, 10034],
    'mott_haven_hunts_point': [10454, 10455, 10474],
    'astoria': [11101, 11102, 11103, 11104, 11105, 11106],
    'sunset_park': [11220, 11232],
    'morrisania_belmont': [14813, 10456, 10459, 10457, 10458],
    'brownsville_ocean_hill': [11212, 11233],
    'crown_heights_south': [11225, 11213],
}
```

```python
non_gentrifying_map = {
    'highbridge_concourse_south': [10451, 10452],
    'kingsbridge_heights_moshulu': [10468, 10463, 10467],
    'university_heights_fordham': [10453, 10468, 10453],
    'soundview_parkchester': [10472, 10473, 10462, 10460, 10461],
    'bensonhurst': [11204, 11214],
    'coney_island': [11224, 11223],
    'east_new_york': [11207, 11208, 11239],
}
```

```python
gentrifying_zipcodes = []
for k,v in gentrifying_map.items():
    for z in v:
        if z not in gentrifying_zipcodes:
            gentrifying_zipcodes.append(z)
```

```python
non_gentrifying_zipcodes = []
for k,v in non_gentrifying_map.items():
    for z in v:
        if z not in non_gentrifying_zipcodes:
            non_gentrifying_zipcodes.append(z)
```

```python
file = 'data/regression_data.csv'
data = pd.read_csv(file)
```

```python
data.head()
```

```python
data['gentrifying'] = 0
data['non_gentrifying'] = 0
data['higher_income'] = 0

for i,r in data.iterrows():
    if r['zipcode'] in gentrifying_zipcodes:
        data.loc[i, 'gentrifying'] = 1
    elif r['zipcode'] in non_gentrifying_zipcodes:
        data.loc[i, 'non_gentrifying'] = 1
    else:
        data.loc[i, 'higher_income'] = 1
```

```python
data.to_csv(file, index=False)
```
