```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn import metrics
from scipy import stats
```

```python
file = '../data/regression_data.csv'
data = pd.read_csv(file)
```

```python
data.head()
```

```python
data['crime_score_norm'] = (data['crime_score']-np.nanmin(data['crime_score']))/(np.nanmax(data['crime_score'])-np.nanmin(data['crime_score']))
```

```python
data['gentrification_score'] = 0

for i,r in data.iterrows():
    data.loc[i, 'gentrification_score'] = np.nanmean([1-r['crime_score_norm'], r['edu_bachelors'], 1-r['race_index'], r['gini_index'], r['age_25_34']])
```

```python
for c in ['gentrifying', 'non_gentrifying', 'higher_income']:
    temp = data[data[c]==1]
    print(np.mean(temp['gentrification_score']))
    print(np.std(temp['gentrification_score']))
    print(np.min(temp['gentrification_score']))
    print(np.max(temp['gentrification_score']))
    print(len(temp))
    print()
```

```python
data.to_csv(file, index=False)
```

```python
# Correlation Analysis

for f in data.columns:
    print(f, stats.pearsonr(data[f], data['gentrification_score']))
```
