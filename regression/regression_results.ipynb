```python
from matplotlib import pyplot as plt
```

```python
def formatFloat(val, fmt="%.2f"):
    ret = fmt % val
    if ret.startswith("0."):
        return ret[1:]
    if ret.startswith("-0."):
        return "-" + ret[2:]
    return ret
```

```python
def formatFloat(val, fmt="%.1f"):
    ret = fmt % val
    if ret.startswith("0."):
        return ret[1:]
    if ret.startswith("-0."):
        return "-" + ret[2:]
    return ret
```

```python
plt.rcParams["figure.figsize"] = [7,5]
groups = ['New York', 'Los Angeles', 'London']
fig, ax = plt.subplots()

index = np.arange(len(groups))
bar_width = 1/5

baseline = [14.696, 15.371, 17.824]
sf_res = [9.159, 12.187, 13.716]
uf_res = [6.322, 6.927, 12.729]
all_res = [5.723, 5.628, 12.087]

# Plot Bars
ax.bar(index, baseline, bar_width, edgecolor='k', color='#C0C0C0', label='Baseline')
ax.bar(index+bar_width, sf_res, bar_width, edgecolor='k', color='#8FBC8F', label='Structured')
ax.bar(index+2*bar_width, uf_res, bar_width, edgecolor='k', color='#bff8c9', label='Unstructured')
ax.bar(index+3*bar_width, all_res, bar_width, edgecolor='k', color='#f1fff7', label='All Variables')

colors = ['#8FBC8F','#bff8c9','#F5FFFA']


# Set Labels
ax.set_ylabel('Root Mean Squared Error (RMSE)',fontsize=14)


for i, v in enumerate(groups):
    ax.text(i-0.095, baseline[i]/2.5, formatFloat(baseline[i]), color='black',fontsize=12)
    if len(formatFloat(sf_res[i]))==4:
        ax.text(i+0.11, sf_res[i]/2.5, formatFloat(sf_res[i]), color='black', fontsize=12)
    else:
        ax.text(i+0.1325, sf_res[i]/2.5, formatFloat(sf_res[i]), color='black', fontsize=12)
    if len(formatFloat(uf_res[i]))==4:
        ax.text(i+2.75*0.1125, uf_res[i]/2.5, formatFloat(uf_res[i]), color='black', fontsize=12)
    else:
        ax.text(i+2.5*0.1325, uf_res[i]/2.5, formatFloat(uf_res[i]), color='black', fontsize=12)
    if len(formatFloat(all_res[i]))==4:
        ax.text(i+4.55*0.1125, all_res[i]/2.5, formatFloat(all_res[i]), color='black', fontsize=12)
    else:
        ax.text(i+4.05*0.1325, all_res[i]/2.5, formatFloat(all_res[i]), color='black', fontsize=12)
        
    #ax.text(i+8*0.07, all_res[i]+1, formatFloat(all_res[i]), color='black', fontsize=12)

#nt[i]+nt_std[i]+0.025
# X-Axis Labels
ax.set_xticks(index+3*bar_width/2)
ax.set_xticklabels(groups,fontsize=14)

# Set Axis Limits
ax.set_ylim(0, 25)
ax.set_yticklabels([0,5,10,15,20,25], fontsize=12)

ax.legend(loc='upper left',fontsize=12)
#plt.title('In-Sample Regression',fontsize=15)
# Helps Things Look Nice When Saving
fig.tight_layout()

# View Plot in Jupyter
plt.show()

# Save Figure (High Quality)
fig.savefig('reg_IS.eps', format='eps', dpi=1000)
```
